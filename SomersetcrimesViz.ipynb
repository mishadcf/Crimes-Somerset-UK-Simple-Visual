```python
df=df.drop(columns=['Context'])
```

```python
df['Crime type']
```

```python
drugs=df.loc[df['Crime type']=='Drugs']
```

```python
drugsloc=drugs[['Latitude', 'Location','Longitude']].dropna()
```

```python
drugsloc
```

```python
centre=[ 51.1800,-3.0000]
```

```python
m = folium.Map(location=centre, zoom_start=9)
```

```python
m
```

```python
# add markers to the map for each crime location
for index, row in drugsloc.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']], 
                  popup=row['Location']).add_to(m)
```
